```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal
from scipy import stats
```

```python
N = int(1e5)
avg_noise = 100e-12 #100 pT/sqrt(Hz)
peak_value = 60000e-12 #60,000 pT/sqrt(Hz)

freq = np.linspace(0,100,N)
baseline = np.full(N,avg_noise)

powerline_peak = peak_value * np.exp(-1*np.abs(freq-60))

background_noise_psd = baseline+powerline_peak
```

```python

plt.plot(freq,background_noise_psd)
```

```python
# Generate a sample signal
N = int(1e7)
t = np.linspace(0, 1, N)
x = np.random.random(N)

# Calculate the power spectral density
frequencies, psd = signal.welch(x)

# Plot the PSD
plt.plot(frequencies, psd)
plt.yscale("log")
plt.xlabel('Frequency [Hz]')
plt.ylabel('Power Spectral Density [V**2/Hz]')
plt.show()
```
